# Understanding Python Collections: Lists, Tuples, and Dictionaries

## 1. Working with Python Lists

Lists are **ordered** and **mutable** collections. You can modify them after creation (add, remove, change elements).

Key characteristics:
- Use square brackets: `[]`
- Elements are indexed starting from 0
- Can contain mixed data types
- Can be modified (add, remove, change items)

```python
# Creating a list
fruits = ["apple", "banana", "cherry"]
print("Fruits list:", fruits)

# Lists can contain mixed data types
mixed_list = [1, "Hello", 3.14, True]
print("Mixed list:", mixed_list)

# Access elements using index (0-based indexing)
print("First fruit:", fruits[0])
print("Last fruit:", fruits[-1])  # Negative indexing from the end
```

```python
# Modifying list elements
numbers = [1, 2, 3, 4, 5]
print("Original list:", numbers)

# Change an element
numbers[0] = 10
print("After changing first element:", numbers)

# Add an element
numbers.append(6)
print("After appending 6:", numbers)

# Remove an element
numbers.remove(3)
print("After removing 3:", numbers)

# Insert at specific position
numbers.insert(2, 99)
print("After inserting 99 at index 2:", numbers)
```

## 2. Working with Python Tuples

Tuples are **ordered** but **immutable** collections. Once created, they cannot be modified.

Key characteristics:
- Use parentheses: `()`
- Elements are indexed starting from 0
- Cannot be changed, added to, or removed after creation
- Often used for fixed data or as dictionary keys (lists cannot be keys)

```python
# Creating tuples
coordinates = (10, 20, 30)
print("Tuple:", coordinates)
print("Type:", type(coordinates))

# Access elements (same as lists)
print("First element:", coordinates[0])
print("Last element:", coordinates[-1])

# Tuples can contain mixed types
student_info = ("Alice", 20, 3.8)
print("Student info:", student_info)
```

```python
# Demonstrating immutability
my_tuple = (1, 2, 3)
print("Original tuple:", my_tuple)

# ❌ This will cause an error - tuples are immutable
try:
    my_tuple[0] = 10  # Attempting to change an element
except TypeError as e:
    print("Error:", e)

# ❌ Cannot use append on tuples
try:
    my_tuple.append(4)
except AttributeError as e:
    print("Error:", e)

print("Tuple remains unchanged:", my_tuple)
```

## 3. Working with Python Dictionaries

Dictionaries store data as **key-value pairs** and are **mutable**.

Key characteristics:
- Use curly braces: `{}`
- Access values using keys (not indexes)
- Keys must be unique and usually strings or numbers
- Excellent for storing structured data (like objects)

```python
# Creating a dictionary
person = {
    "name": "John",
    "age": 25,
    "city": "New York",
    "occupation": "Engineer"
}
print("Person dictionary:", person)

# Access values using keys
print("Name:", person["name"])
print("Age:", person["age"])

# Another example: student records
student = {
    "student_id": "S001",
    "name": "Emma",
    "gpa": 3.9,
    "major": "Computer Science"
}
print("\nStudent info:", student)
print("Student name:", student["name"])
print("GPA:", student["gpa"])
```

```python
# Modifying dictionaries
employee = {"name": "Michael", "salary": 50000, "department": "Sales"}
print("Original employee:", employee)

# Update a value
employee["salary"] = 55000
print("After raise:", employee)

# Add a new key-value pair
employee["years_experience"] = 5
print("After adding years_experience:", employee)

# Remove a key-value pair
del employee["department"]
print("After removing department:", employee)

# Check if a key exists
if "salary" in employee:
    print("Salary is in employee data")
```

## 4. Choosing the Right Data Structure

### Quick Comparison

| Feature | List | Tuple | Dictionary |
|---------|------|-------|-----------|
| **Syntax** | `[]` | `()` | `{}` |
| **Mutable** | ✅ Yes | ❌ No | ✅ Yes |
| **Ordered** | ✅ Yes | ✅ Yes | ✅ Yes (Python 3.7+) |
| **Access** | By index | By index | By key |
| **Use Case** | Collections that change | Fixed data, constants | Structured data, records |
| **Performance** | Good | Faster | Good for lookups |

### When to Use Each

**Use Lists when:**
- You have a collection of similar items that may change
- Order matters (students in a class, steps in a recipe)
- You need to add, remove, or modify items

**Use Tuples when:**
- Data should never change (immutable protection)
- Using fixed, known values (coordinates, RGB colors)
- Using as dictionary keys or in sets (can't use lists for this)

**Use Dictionaries when:**
- You need to associate related data (label each value)
- You want to look up values by meaningful names (not just indexes)
- Modeling objects or records (like database rows)

```python
# Practical Examples

# Example 1: Storing a list of items that may change
shopping_list = ["milk", "eggs", "bread", "butter"]
shopping_list.append("cheese")
print("Shopping list (list):", shopping_list)

# Example 2: Fixed RGB color - should never change
red_color = (255, 0, 0)  # (R, G, B) - tuple is better for fixed data
print("Red color (tuple):", red_color)

# Example 3: Student record with related data
student_record = {
    "id": "S12345",
    "name": "Sarah",
    "email": "sarah@school.edu",
    "gpa": 3.7,
    "enrolled_courses": ["Math", "Physics", "Chemistry"]  # List inside dictionary
}
print("Student record (dictionary):", student_record)
print("Student name:", student_record["name"])
print("Courses:", student_record["enrolled_courses"])
```

### Why This Matters

Choosing the right collection prevents bugs, makes code readable, and ensures data integrity. These three structures handle 90% of data storage needs in Python.
