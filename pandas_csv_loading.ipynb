## 4. Understanding CSV Structure

- Each row represents a record
- Each column represents a feature
- The first row contains column names (headers)
- Data is separated using commas

## 5. Common Loading Issues

- Incorrect file path
- Missing headers
- Extra or missing columns
- Data not aligned properly

Always inspect data after loading.

## Conclusion

CSV files can be loaded into Pandas DataFrames using read_csv(). After loading, it is important to inspect the structure, column names, and data types to ensure correctness before performing any analysis.
