# Selecting Rows and Columns Using Indexing and Slicing (Milestone 5.32)

This notebook demonstrates safe and intentional data selection in Pandas.

You will practice:
- Selecting columns by name
- Selecting rows by position with `iloc`
- Selecting rows by label with `loc`
- Slicing rows and columns together
- Avoiding common selection mistakes

```python
import pandas as pd
```

## 1. Load a DataFrame

We will use the employee attrition dataset from the workspace.

```python
csv_path = 'employee-survey-analysis/data/raw/Employee Attrition.csv'
df = pd.read_csv(csv_path)

print('Shape:', df.shape)
print('First 5 columns:', df.columns[:5].tolist())
df.head()
```

## 2. Selecting Columns by Name

Single column returns a **Series**.
Multiple columns return a **DataFrame**.

```python
# Single column (Series)
age_col = df['Age']
print(type(age_col))
print(age_col.head())
```

```python
# Multiple columns (DataFrame)
basic_cols = df[['Age', 'Department', 'JobRole']]
print(type(basic_cols))
basic_cols.head()
```

## 3. Selecting Rows by Position with iloc

`iloc` uses zero-based integer positions.

```python
# First row (position 0)
row0 = df.iloc[0]
row0
```

```python
# Rows from position 5 to 9 (10 is excluded)
rows_5_to_9 = df.iloc[5:10]
rows_5_to_9
```

## 4. Selecting Rows by Label with loc

`loc` uses index labels. To make label-based selection clear, we set a named index.

```python
df_labeled = df.copy()
df_labeled.index = [f'EMP_{i:04d}' for i in range(1, len(df_labeled) + 1)]
df_labeled.head()
```

```python
# Single row by label
df_labeled.loc['EMP_0001']
```

```python
# Label slicing is inclusive for both ends
df_labeled.loc['EMP_0005':'EMP_0010']
```

## 5. Selecting Rows and Columns Together

Use `loc` or `iloc` to combine row and column selection in one clear expression.

```python
# By position: rows 0-4 and selected column positions
subset_iloc = df.iloc[0:5, [0, 2, 4]]
subset_iloc
```

```python
# By label: specific rows and columns
subset_loc = df_labeled.loc['EMP_0003':'EMP_0008', ['Age', 'Department', 'MonthlyIncome']]
subset_loc
```

## 6. Common Mistakes to Avoid

- Do not confuse `iloc` (position) with `loc` (label).
- Avoid chained indexing for assignment (for example, `df[df['Age'] > 30]['Age'] = 0`).
- Verify the result shape after each selection using `.shape`.

Quick checks:

```python
print('subset_iloc shape:', subset_iloc.shape)
print('subset_loc shape:', subset_loc.shape)
print('subset_loc columns:', subset_loc.columns.tolist())
```
