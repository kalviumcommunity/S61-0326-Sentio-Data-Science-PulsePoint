```python
import pandas as pd

# 1. Load a DataFrame
# Use the correct relative path for the dataset
csv_path = '../data/raw/Employee Attrition.csv'
df = pd.read_csv(csv_path)

# 2. Inspect DataFrame Shape
print('Shape:', df.shape)
print(f'Rows (observations): {df.shape[0]}')
print(f'Columns (features): {df.shape[1]}')

# 3. Inspect Column Data Types
print('\nColumn Data Types:')
print(df.dtypes)

# 4. Detect Type-Related Issues
# Find columns with object (string) type that may be numeric
object_cols = df.select_dtypes(include='object').columns
for col in object_cols:
    try:
        pd.to_numeric(df[col])
        print(f'Column "{col}" looks numeric but is stored as object (string).')
    except Exception:
        pass

# Check for missing values affecting types
print('\nMissing values per column:')
print(df.isnull().sum())
```

# Identifying and Removing Duplicate Records

Duplicate records can distort your analysis and lead to incorrect conclusions. This section will guide you through understanding, detecting, removing, and verifying duplicates in a Pandas DataFrame.

## 1. Understanding Duplicate Records

- **Duplicate rows** are records that appear more than once in your dataset.
- Duplicates can be exact (all columns match) or partial (some columns match).
- Causes include data entry errors, merging datasets, or system glitches.
- Always inspect your data before assuming uniqueness.

## 2. Detecting Duplicate Rows

Let's find duplicate rows in the DataFrame. We'll use `duplicated()` to get a boolean indicator and `sum()` to count them.

```python
# Detect duplicate rows (all columns)
duplicated_mask = df.duplicated()
num_duplicates = duplicated_mask.sum()
print(f"Number of duplicate rows: {num_duplicates}")

# Show duplicate rows (excluding the first occurrence)
duplicates = df[duplicated_mask]
print("\nDuplicate rows:")
display(duplicates.head())  # Use display() in Jupyter for better formatting
```

### Detecting Duplicates on Selected Columns

You can check for duplicates based on specific columns (e.g., 'EmployeeNumber', 'EmployeeID', or a combination):

- Use `df.duplicated(subset=[...])` to specify columns.

```python
# Example: Check duplicates based on 'EmployeeNumber' (replace with relevant column)
if 'EmployeeNumber' in df.columns:
    empnum_duplicates = df.duplicated(subset=['EmployeeNumber'])
    print(f"Duplicate EmployeeNumber rows: {empnum_duplicates.sum()}")
    display(df[empnum_duplicates].head())
```

## 3. Removing Duplicate Records

Once duplicates are detected, you can remove them using `drop_duplicates()`. By default, it keeps the first occurrence and removes subsequent ones.

```python
# Remove duplicate rows (all columns), keeping the first occurrence
df_dedup = df.drop_duplicates()
print(f"Shape after removing duplicates: {df_dedup.shape}")
```

```python
# Remove duplicates based on selected columns (e.g., 'EmployeeNumber')
if 'EmployeeNumber' in df.columns:
    df_empnum_dedup = df.drop_duplicates(subset=['EmployeeNumber'])
    print(f"Shape after removing EmployeeNumber duplicates: {df_empnum_dedup.shape}")
```

## 4. Verifying Deduplication Results

After removing duplicates, always verify:
- The new shape of your DataFrame
- That no duplicates remain
- That important records are preserved

```python
# Check for remaining duplicates in the deduplicated DataFrame
remaining_duplicates = df_dedup.duplicated().sum()
print(f"Remaining duplicate rows after deduplication: {remaining_duplicates}")

# Compare shapes before and after
def print_shape_comparison(original, deduped):
    print(f"Original shape: {original.shape}")
    print(f"After deduplication: {deduped.shape}")

print_shape_comparison(df, df_dedup)
```
