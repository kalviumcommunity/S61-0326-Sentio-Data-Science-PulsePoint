```python
# Pandas Fundamentals: Loading and Inspecting CSV Data

# This notebook will guide you through the essential steps of loading a CSV file into a Pandas DataFrame, inspecting its structure, and recognizing common issues that can occur during the loading process.

# 1. What is a CSV File?
# - CSV stands for Comma-Separated Values.
# - It is a plain text format for storing tabular data (like a spreadsheet).
# - Each line is a row; values are separated by commas (or another delimiter).
# - The first row often contains column headers.

# 2. Loading a CSV File into Pandas
# Let's load the Employee Attrition dataset.
import pandas as pd
from pathlib import Path

# Use absolute path based on notebook location
file_path = Path.cwd().parent / "data" / "raw" / "Employee Attrition.csv"
df = pd.read_csv(file_path)

# 3. Inspecting the Loaded Data
# Preview the first few rows:
print(df.head())

# Check column names:
print(df.columns)

# Check the number of rows and columns:
print(df.shape)  # (rows, columns)

# Get a summary of the DataFrame:
print(df.info())

# 4. Recognizing Common Loading Issues
# - Unexpected column names: Check for extra delimiters or encoding issues.
# - Data shifted into wrong columns: Look for missing/extra commas in the CSV.
# - If the first row is not headers, use `header=None` in `read_csv()`.
# - If the delimiter is not a comma, specify it: `pd.read_csv(file_path, delimiter=';')`

# **Always inspect your DataFrame after loading to catch issues early!**
```
