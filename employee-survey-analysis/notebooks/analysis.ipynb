# Code Cells and Markdown Cells in Jupyter Notebook

This notebook demonstrates how to use **Code** cells for execution and **Markdown** cells for explanation.

## Purpose of This Notebook

This notebook is part of a fundamentals milestone.

- Code cells are used to run Python
- Markdown cells are used to explain what the notebook is doing
- Keeping them separate makes notebooks easier to read and review

## Example 1: A Simple Code Cell

The next cell runs a basic Python command. This is a **Code** cell because it performs execution and produces output.

```python
print("Hello from a Code cell!")
```

## Example 2: Another Code Cell

This next cell stores values in variables and performs a small calculation. The explanation belongs here in Markdown, while the computation belongs in the Code cell below.

```python
a = 7
b = 8
total = a + b
print("The sum is:", total)
```

## When to Use Each Cell Type

**Use a Code cell when you want to:**

- Run Python
- Create variables
- Perform calculations
- Display output

**Use a Markdown cell when you want to:**

- Add a title or section heading
- Explain what the next code cell does
- Describe results in words
- Make the notebook readable for other people

## Practicing Cell Switching

In Jupyter Notebook, you can change a cell type from the toolbar or by using command mode:

- Press `Esc` then `M` to convert a cell to Markdown
- Press `Esc` then `Y` to convert a cell to Code

This helps fix mistakes and keeps the notebook organized.
