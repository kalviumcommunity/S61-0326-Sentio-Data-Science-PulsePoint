# Understanding Python Numeric and String Data Types

## 1. Working with Numeric Data Types

```python
# Integer
a = 10
b = 5

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
```

```python
# Float
x = 10.5
y = 2.5

print("Division:", x / y)
print("Floor Division:", x // y)
```

## 2. Understanding String Data Types

```python
# Strings
name = "Ishana"
message = "Learning Python"

print(name)
print(message)
```

```python
# Concatenation
full_message = name + " is " + message
print(full_message)
```

## 3. Mixing Numbers and Strings Safely

```python
age = 20

# ❌ This will cause error
# print("Age is " + age)

# ✅ Correct way
print("Age is " + str(age))
```

```python
# Convert string to number
num = "15"
print(int(num) + 5)
```

## 4. Inspecting Data Types

```python
a = 10
b = 10.5
c = "Hello"

print(type(a))
print(type(b))
print(type(c))
```

## Conclusion

In this notebook, basic Python data types such as integers, floats, and strings were explored. Simple operations were performed to understand how Python handles numbers and text. Additionally, type conversion and type checking were used to avoid errors and ensure correct program behavior.

This forms the foundation for working with real-world data in future tasks.
