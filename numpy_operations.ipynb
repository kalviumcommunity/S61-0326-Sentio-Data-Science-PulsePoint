# Performing Basic Mathematical Operations on NumPy Arrays

## 1. Importing NumPy

```python
import numpy as np
```

## 2. Creating Arrays

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

print("Array 1:", arr1)
print("Array 2:", arr2)
```

## 3. Element-Wise Operations

```python
# Addition
print("Addition:", arr1 + arr2)

# Subtraction
print("Subtraction:", arr1 - arr2)

# Multiplication
print("Multiplication:", arr1 * arr2)

# Division
print("Division:", arr1 / arr2)
```

## 4. Scalar Operations

```python
# Add scalar
print("Add 2:", arr1 + 2)

# Multiply scalar
print("Multiply by 3:", arr1 * 3)
```

## 5. NumPy vs Python Lists

```python
# Python list behavior
list1 = [1, 2, 3]

print("List * 2:", list1 * 2)   # duplicates
```

```python
# NumPy array behavior
arr = np.array([1, 2, 3])

print("Array * 2:", arr * 2)   # multiplies
```

## 6. Shape Compatibility

```python
a = np.array([1, 2, 3])
b = np.array([1, 2])

# This will cause an error if uncommented
# print(a + b)
```

## Conclusion

NumPy allows mathematical operations to be performed element-wise on arrays, making computations faster and more efficient compared to Python lists. Scalar operations apply to all elements, and understanding shape compatibility is important to avoid errors. These operations form the foundation for numerical computing in data science.
