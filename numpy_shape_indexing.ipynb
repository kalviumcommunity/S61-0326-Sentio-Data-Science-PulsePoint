# Understanding Array Shape, Dimensions, and Index Positions

## 1. Understanding Array Shape

```python
import numpy as np

# 1D array
arr1 = np.array([10, 20, 30, 40])

print("Array:", arr1)
print("Shape:", arr1.shape)
```

```python
# 2D array
arr2 = np.array([[1, 2, 3],
                 [4, 5, 6]])

print("Array:\n", arr2)
print("Shape:", arr2.shape)
```

## 2. Understanding Dimensions (ndim)

```python
print("arr1 dimensions:", arr1.ndim)
print("arr2 dimensions:", arr2.ndim)
```

## 3. Accessing Elements Using Index Positions

```python
# Access elements
print("First element:", arr1[0])
print("Second element:", arr1[1])
```

```python
# Access elements in 2D
print("Element at row 0, col 1:", arr2[0][1])
print("Element at row 1, col 2:", arr2[1][2])
```

## 4. Visualizing Array Layout

2D Array:

[[1, 2, 3],
 [4, 5, 6]]

Index positions:

(0,0) → 1  
(0,1) → 2  
(0,2) → 3  
(1,0) → 4  
(1,1) → 5  
(1,2) → 6  

Rows come first, then columns.

```python
# Example of valid indexing
print(arr2[1][1])  # 5

# Invalid (will cause error if uncommented)
# print(arr2[2][0])
```

## Conclusion

Array shape defines how data is organized, while dimensions describe the complexity of the array. Proper indexing allows safe access to elements. Understanding these concepts helps avoid errors and builds a strong foundation for further NumPy operations.
