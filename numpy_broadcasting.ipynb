# Understanding NumPy Broadcasting with Simple Examples

## 1. Importing NumPy

```python
import numpy as np
```

## 2. Broadcasting with Scalars

```python
arr = np.array([1, 2, 3])

print("Original Array:", arr)
print("Add 5:", arr + 5)
print("Multiply by 2:", arr * 2)
```

A scalar is applied to every element of the array. This is the simplest form of broadcasting.

## 3. Broadcasting Between 1D Arrays

```python
a = np.array([1, 2, 3])
b = np.array([10, 20, 30])

print("Addition:", a + b)
```

Arrays of the same shape are combined element-wise.

## 4. Broadcasting Between 2D and 1D Arrays

```python
matrix = np.array([[1, 2, 3],
                   [4, 5, 6]])

vector = np.array([10, 20, 30])

print("Result:\n", matrix + vector)
```

The 1D array is applied across each row of the 2D array.

## 5. Understanding Shapes

```python
print("Matrix shape:", matrix.shape)
print("Vector shape:", vector.shape)
```

## 6. Incompatible Shapes

```python
x = np.array([1, 2, 3])
y = np.array([1, 2])

# This will cause an error if uncommented
# print(x + y)
```

Broadcasting fails when shapes are not compatible.

## 7. Broadcasting Rules (Simple)

- NumPy compares shapes from right to left
- Dimensions must match OR one of them must be 1
- If not, broadcasting fails

## Conclusion

Broadcasting allows NumPy to perform operations on arrays of different shapes without using loops. Scalars apply to all elements, and smaller arrays are stretched to match larger ones when compatible. Understanding shapes helps avoid errors and ensures correct results.
