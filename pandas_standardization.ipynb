# Standardizing Column Names and Data Formats

## 1. Importing Pandas

import pandas as pd

## 2. Loading Dataset

df = pd.read_csv("data/employee_data.csv")
df

## 3. Standardizing Column Names

Convert column names to lowercase, remove spaces, and replace with underscores.

df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

print(df.columns)

## 4. Standardizing Text Data

Convert text to lowercase and remove extra spaces.

df["feedback_text"] = df["feedback_text"].str.lower().str.strip()
df["department_name"] = df["department_name"].str.lower().str.strip()

df

## 5. Standardizing Numeric Data

Ensure numeric columns are correctly formatted.

df["rating"] = pd.to_numeric(df["rating"], errors="coerce")

df.info()

## 6. Standardizing Date Format

df["date"] = pd.to_datetime(df["date"], errors="coerce")

df

## 7. Naming Convention

- All column names are in lowercase
- Words are separated using underscores (snake_case)
- No spaces or special characters

## Conclusion

Standardizing column names and data formats improves readability and consistency. Clean column names make coding easier, and standardized data formats ensure reliable processing and analysis.
