# Pandas DataFrame Fundamentals

This notebook demonstrates how to create Pandas DataFrames from dictionaries and files, and how to inspect their structure and contents. DataFrames are the backbone of data analysis in Python.

## 1. Import Pandas Library

First, import the pandas library using the standard alias `pd`.

```python
import pandas as pd
```

## 2. Understanding Pandas DataFrames

A Pandas DataFrame is a two-dimensional, tabular data structure with labeled rows and columns. It is similar to a spreadsheet or a SQL table.

- **Rows**: Represent individual records (like spreadsheet rows)
- **Columns**: Represent variables or features (like spreadsheet columns)
- **Column names**: Labels for each column
- **Index**: Labels for each row (default is integer-based)

Below is a simple example:

```python
# Example DataFrame structure
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Paris', 'London']
}
df_example = pd.DataFrame(data)
df_example
```

## 3. Creating DataFrames from Dictionaries

You can create a DataFrame directly from a Python dictionary, where each key is a column name and each value is a list of column values.

```python
# Create a dictionary with column-style data
data_dict = {
    'Student': ['John', 'Emma', 'Lucas'],
    'Score': [88, 92, 79],
    'Passed': [True, True, False]
}

# Convert the dictionary into a DataFrame
df_dict = pd.DataFrame(data_dict)
df_dict
```

```python
# Inspect column names and values
print('Columns:', df_dict.columns.tolist())
print('Values:')
print(df_dict.values)
```

## 4. Creating DataFrames from Files

You can also create a DataFrame by loading data from a file, such as a CSV. This is the most common way to work with real-world data.

```python
# Load a DataFrame from a CSV file
# (Using the sample Employee Attrition.csv from the workspace)
csv_path = 'employee-survey-analysis/data/raw/Employee Attrition.csv'
df_file = pd.read_csv(csv_path)
df_file.head()
```

## 5. Inspecting DataFrame Structure and Contents

After loading data, always inspect the DataFrame to understand its structure and catch any issues early.

```python
# View the first few rows
print('First 5 rows:')
print(df_file.head())

# Inspect column names
print('Columns:', df_file.columns.tolist())

# Check shape (rows, columns)
print('Shape:', df_file.shape)

# Check data types
print('Data types:')
print(df_file.dtypes)
```
