The head() method shows the first few rows of the dataset. It helps quickly understand column names and sample values.

## 4. Inspecting Structure with info()

The info() method provides information about:
- Column names
- Data types
- Non-null values
- Memory usage

## 5. Summarizing Data with describe()

The describe() method gives statistical summary of numeric columns such as:
- Count
- Mean
- Minimum and Maximum
- Percentiles

## 6. When to Use Each Method

- head() → To preview data
- info() → To understand structure and data types
- describe() → To analyze numeric statistics

Using all three together gives a complete understanding of the dataset.

## Conclusion

Data inspection is a critical first step in any data science workflow. Using head(), info(), and describe() helps understand the dataset structure, identify issues early, and ensures that further analysis is based on correct data.
