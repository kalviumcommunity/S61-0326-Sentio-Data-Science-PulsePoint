# Passing Data into Functions and Returning Results

This notebook demonstrates how to pass data into functions and return results in Python.

## 1. Understanding Parameters and Arguments

Functions accept inputs using parameters, and we pass values using arguments.

```python
# Function with parameters
def greet(name):
    return "Hello, " + name

# Calling function
result = greet("Ishana")
print(result)
```

## 2. Returning Values from Functions

Functions should return values instead of printing directly.

```python
def add_numbers(a, b):
    return a + b

sum_result = add_numbers(5, 10)
print("Sum:", sum_result)
```

## 3. Using Returned Results

Returned values can be reused in further calculations.

```python
def multiply(x, y):
    return x * y

def square(n):
    return n * n

result = multiply(3, 4)
final_result = square(result)

print("Final Result:", final_result)
```

## 4. Avoiding Common Function Mistakes

Avoid printing inside functions when return is needed.

```python
def wrong_function(x):
    print(x * 2)
```

```python
def correct_function(x):
    return x * 2

value = correct_function(5)
print(value)
```
