# Using for and while Loops for Iterative Data Processing

## 1. Using for Loops for Iteration

```python
# Loop through numbers
for i in range(1, 6):
    print("Number:", i)
```

```python
# Loop through a list
fruits = ["apple", "banana", "cherry"]

for fruit in fruits:
    print("Fruit:", fruit)
```

## 2. Using while Loops for Condition-Based Repetition

```python
count = 1

while count <= 5:
    print("Count:", count)
    count += 1
```

## 3. Controlling Loop Flow

```python
for i in range(1, 10):
    if i == 5:
        break
    print(i)
```

```python
for i in range(1, 6):
    if i == 3:
        continue
    print(i)
```

## 4. Avoiding Infinite Loops

```python
# Correct loop (terminates)
x = 1

while x <= 3:
    print("Value:", x)
    x += 1
```

### Important Note:
A loop becomes infinite if the condition never becomes false.
Always update the loop variable.

## Conclusion

In this notebook, for and while loops were used to perform iterative operations. The for loop was used for fixed sequences, while the while loop was used for condition-based repetition. Loop control statements like break and continue were used to manage execution. Proper handling of loop conditions helps avoid infinite loops and ensures correct program behavior.
